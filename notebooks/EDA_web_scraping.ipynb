```python
import yaml

# Sample markdown content
markdown_content = documents[1].get("content")

# Initialize the parser with the markdown content and configuration
parser = MarkdownParser(content=markdown_content)
```

```python
# Summarize the structure
structure_summary = parser.summarize_structure()
print("Structure Summary:")
print(yaml.dump(structure_summary, sort_keys=False))
```

```python
documents[1].get("metadata").get("urls")
```

```python
# Extract plain text with associations
plain_text_associated = parser.extract_plain_text_with_associations()
print("\nPlain Text with Associations:")
for elem in plain_text_associated:
    print(f"Type: {elem['type']}, Parent: {elem['parent']}, Content: {elem['content']}")
```

```python
unique_ordered = list(
    dict.fromkeys(
        elem.get("parent")
        for elem in plain_text_associated
        if elem.get("parent") is not None
    )
)
unique_ordered
```

```python

# Extract all links
links = parser.extract_links()
print("\nExtracted Links:")
for link in links:
    print(link)
```

```python
# Extract all images
images = parser.extract_images()
print("\nExtracted Images:")
for img in images:
    print(img)
```

```python
# Generate Table of Contents
toc = parser.generate_toc()
print("\nTable of Contents:")
print(toc)
```

```python
# Extract plain text without associations
plain_text = parser.extract_plain_text()
print("\nPlain Text:")
print(plain_text)
```

```python
# Get comprehensive statistics
stats = parser.get_statistics()
print("\nComprehensive Statistics:")
print(yaml.dump(stats, sort_keys=False))
```
